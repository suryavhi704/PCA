## **REQUIRED LIBRARIES:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

## **DATA READING TECHNIQUE:-**

```python
df=pd.read_csv('customer_churn_ (1).csv')
df.info()
```

```python
df.head(6)
```

## **EDA PROCESES:-**

```python
df.isnull().sum() # checking of null values
```

```python
df.duplicated().sum() # checking of duplicated values
```

```python
df.drop(columns='customerID',inplace=True) # dropping of customerID column
```

## **LABEL ENCODING:-**

```python
obj_columns=df.select_dtypes('object').columns
obj_columns
```

```python
encoder=LabelEncoder()
for col in obj_columns:
    df[col]=encoder.fit_transform(df[col])
```

```python
df
```

## **standerdization of required column:-**

```python
sc=StandardScaler()
df['TotalCharges']=sc.fit_transform(df[['TotalCharges']])
```

```python
df
```

## **ML PROCESS:-**

```python
X=df.drop(columns='Churn')
y=df['Churn']
```

```python
# check for imbalance data
y.value_counts()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

## **PCA PROCESS & MODEL BUILDING PROCESS::-**

## **LOGISTIC REGRESSION:-**

```python
component_list=[5,10,15,18]
for component in component_list:
  pca=PCA(n_components=component)
  x_train_pca=pca.fit_transform(x_train)
  x_test_pca=pca.transform(x_test)

  model1=LogisticRegression()
  model1.fit(x_train_pca,y_train)
  y_pred=model1.predict(x_test_pca)
  print(f'roc_auc score for {component} components is:{roc_auc_score(y_test,y_pred)}')
  print(f'Accuracy score for {component} components is:{accuracy_score(y_test,y_pred)}\n')
```

## **DECISION TREE:-**

```python
component_list=[5,10,15,18]
for component in component_list:
  pca=PCA(n_components=component)
  x_train_pca=pca.fit_transform(x_train)
  x_test_pca=pca.transform(x_test)

  model2=DecisionTreeClassifier()
  model2.fit(x_train_pca,y_train)
  y_pred=model2.predict(x_test_pca)
  print(f'roc_auc score for {component} components is:{roc_auc_score(y_test,y_pred)}')
  print(f'Accuracy score for {component} components is:{accuracy_score(y_test,y_pred)}\n')
```

## **RANDOM FOREST:-**

```python
component_list=[5,10,15,18]
for component in component_list:
  pca=PCA(n_components=component)
  x_train_pca=pca.fit_transform(x_train)
  x_test_pca=pca.transform(x_test)

  model3=RandomForestClassifier(n_estimators=100,random_state=42)
  model3.fit(x_train_pca,y_train)
  y_pred=model3.predict(x_test_pca)
  print(f'roc_auc score for {component} components is:{roc_auc_score(y_test,y_pred)}')
  print(f'Accuracy score for {component} components is:{accuracy_score(y_test,y_pred)}\n')
```
